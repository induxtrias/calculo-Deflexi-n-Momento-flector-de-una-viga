```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir las propiedades de la viga
L = 5  # Longitud de la viga en metros
E = 200e9  # Módulo de elasticidad del acero en Pa
I = 1e-5  # Momento de inercia de la sección transversal en m^4
P = 10000  # Carga puntual en N
h = 0.2  # Altura de la sección transversal en metros
b = 0.1  # Ancho de la sección transversal en metros

# Crear un array de puntos a lo largo de la viga
x = np.linspace(0, L, 100)

# Calcular la deflexión de la viga
def deflexion(x):
    return (P * x**2 * (3*L - x)) / (6 * E * I)

# Calcular el momento flector
def momento_flector(x):
    return P * (L - x)

# Calcular la fuerza cortante
def fuerza_cortante(x):
    return -P * np.ones_like(x)

# Calcular los resultados
y = deflexion(x)
M = momento_flector(x)
V = fuerza_cortante(x)

# Crear las gráficas
fig = plt.figure(figsize=(15, 15))

# Gráfica de deflexión
ax1 = fig.add_subplot(321)
ax1.plot(x, y)
ax1.set_title('Deflexión de la viga')
ax1.set_xlabel('Posición (m)')
ax1.set_ylabel('Deflexión (m)')

# Gráfica de momento flector
ax2 = fig.add_subplot(323)
ax2.plot(x, M)
ax2.set_title('Diagrama de momento flector')
ax2.set_xlabel('Posición (m)')
ax2.set_ylabel('Momento (N·m)')

# Gráfica de fuerza cortante
ax3 = fig.add_subplot(325)
ax3.plot(x, V)
ax3.set_title('Diagrama de fuerza cortante')
ax3.set_xlabel('Posición (m)')
ax3.set_ylabel('Fuerza cortante (N)')

# Dibujo de la viga en perfil
ax4 = fig.add_subplot(322)
ax4.plot([0, L], [0, 0], 'k-', linewidth=2)
ax4.plot([L/2], [0], 'rv', markersize=10)  # Carga puntual
ax4.text(L/2, -0.02, f'{P/1000} kN', ha='center')
ax4.set_ylim(-0.1, 0.1)
ax4.set_aspect('equal')
ax4.set_title('Perfil de la viga')
ax4.set_xlabel('Longitud (m)')
ax4.axis('off')

# Dibujo de la viga en alzado
ax5 = fig.add_subplot(324)
ax5.plot([0, 0], [-h/2, h/2], 'k-', linewidth=2)
ax5.plot([L, L], [-h/2, h/2], 'k-', linewidth=2)
ax5.plot([0, L], [h/2, h/2], 'k-', linewidth=2)
ax5.plot([0, L], [-h/2, -h/2], 'k-', linewidth=2)
ax5.set_aspect('equal')
ax5.set_title('Alzado de la viga')
ax5.set_xlabel('Longitud (m)')
ax5.set_ylabel('Altura (m)')

# Dibujo de la viga en planta
ax6 = fig.add_subplot(326)
ax6.plot([0, L], [b/2, b/2], 'k-', linewidth=2)
ax6.plot([0, L], [-b/2, -b/2], 'k-', linewidth=2)
ax6.plot([0, 0], [-b/2, b/2], 'k-', linewidth=2)
ax6.plot([L, L], [-b/2, b/2], 'k-', linewidth=2)
ax6.set_aspect('equal')
ax6.set_title('Planta de la viga')
ax6.set_xlabel('Longitud (m)')
ax6.set_ylabel('Ancho (m)')

plt.tight_layout()
plt.show()

# Imprimir resultados máximos
print(f"Deflexión máxima: {np.max(np.abs(y)):.6f} m")
print(f"Momento flector máximo: {np.max(np.abs(M)):.2f} N·m")
print(f"Fuerza cortante máxima: {np.max(np.abs(V)):.2f} N")
```
